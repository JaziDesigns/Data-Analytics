### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Datasets

**Netflix stocks dataset**

```python
netflix_stocks = pd.read_csv("NFLX.csv")
netflix_stocks.sample(5)
```

```python
# Data Information
netflix_stocks.info()
```

```python
# Univariate Analysis
netflix_stocks.describe()
```

```python
# Bivariate Analysis
plt.figure(figsize = (18, 7))
correlation = netflix_stocks.corr()
sns.heatmap(correlation, annot = True)
plt.title("Correlation Matrix")
plt.show()
```

**Dowjones stocks dataset**

```python
dowjones_stocks = pd.read_csv("DJI.csv")
dowjones_stocks.sample(5)
```

```python
# Data Information
dowjones_stocks.info()
```

```python
# Univariate Analysis
dowjones_stocks.describe()
```

```python
# Bivariate Analysis
plt.figure(figsize = (18, 7))
correlation = dowjones_stocks.corr()
sns.heatmap(correlation, annot = True, cmap="Greens_r")
plt.title("Correlation Matrix")
plt.show()
```

**Rename Adj Close to Price**

```python
netflix_stocks.rename(columns = {"Adj Close": "Price"}, inplace = True)
dowjones_stocks.rename(columns = {"Adj Close": "Price"}, inplace = True)
```

**Visualize the distribution of the Netflix stock prices to see the range of prices for 2017**

```python
plt.figure(figsize = (18, 7))
sns.violinplot(x = netflix_stocks['Price'], data = netflix_stocks, color="red")
plt.title("Distribution of 2017 Netflix Stock Price", size = 15)
plt.xlabel("Price", size = 15)
plt.show()
```

**We will chart the performance of the earnings per share (EPS) by graphing the estimate Yahoo projected for the Quarter compared to the actual earnings for that quarters**

```python
x_positions = [1, 2, 3, 4]
chart_labels = ["1Q2017","2Q2017","3Q2017","4Q2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]
plt.figure(figsize = (18, 7))
plt.scatter(x_positions, earnings_actual,  color="red",alpha=0.5, s = 100)
plt.scatter(x_positions, earnings_estimate, color="blue",alpha=0.5, s = 100)
plt.xticks(x_positions, chart_labels)
plt.legend(["Actual", "Estimate"])
plt.title("Earnings Per Share in Cents", size = 15)
plt.show()
```

**We will visualize Netflix company revenue and earnings by mapping two bars side-by-side**

```python
# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q2017","3Q2017","4Q2017", "1Q2018"]
df = pd.DataFrame({"Revenue by Quarter": revenue_by_quarter,
                   "Earnings by Quarter": earnings_by_quarter,
                   "Quarter Labels": quarter_labels})
df["% of Revenue constitute Earnings"] = ((df["Earnings by Quarter"] / df["Revenue by Quarter"]) * 100).round(2)
df
```

```python
# Melt
df = df[["Revenue by Quarter","Earnings by Quarter", "Quarter Labels"]]
df = df.melt("Quarter Labels", var_name = "Revenue & Earnings", value_name = "Amount")
df
```

```python
plt.figure(figsize = (18, 7))
graph = sns.barplot(x = "Quarter Labels", y = "Amount", hue = "Revenue & Earnings", data = df, palette = "plasma")
for p in graph.patches:
        graph.annotate('{:.01f}'.format(p.get_height()), 
                      (p.get_x()+0.2, p.get_height()),
                       ha='center', va='bottom',color= 'black', size = 15)
plt.title("Revenues and Earnings by Quarter", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (18, 7))
fig.suptitle("Comparison of Netflix stock to the Dow Jones Industrial Average in 2017", fontsize = 20)

# Netflix Graph
sns.lineplot(ax = axes[0], x = "Date", y = "Price", data = netflix_stocks, color = "red")
axes[0].set_title("Netflix Stock Price by Dates", size = 15)
axes[0].set_xlabel("Date", size = 15)
axes[0].set_ylabel("Netflix Stock Price", size = 15)
axes[0].tick_params(axis='x', labelrotation=90)

# Dowjones Graph
sns.lineplot(ax = axes[1], x = "Date", y = "Price", data = dowjones_stocks, color = "blue")
axes[1].set_title("Dowjones Stock Price by Dates", size = 15)
axes[1].set_xlabel("Date", size = 15)
axes[1].set_ylabel("Dowjones Stock Price", size = 15)
axes[1].tick_params(axis='x', labelrotation=90)

plt.show()
```
